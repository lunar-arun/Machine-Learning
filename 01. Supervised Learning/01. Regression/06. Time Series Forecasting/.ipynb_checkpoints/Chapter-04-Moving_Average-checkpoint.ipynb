    Moving average (MA) process, 
        the current value depends linearly on the mean of the series, the current error term, and past error terms.
        
        denoted as MA(q), where q is the order.
        General Equation : yt = μ + ϵt + θ1ϵt–1 + θ2ϵt–2 +⋅⋅⋅+ θqϵt–q

    Test for stationarity

```python
from statsmodels.tsa.stattools import adfuller

ADF_result = adfuller(df['widget_sales'])

print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
```

    Inference:
        ADF statistic is not a large negative number, and the p-value is greater than 0.05.
        
        Hence : time series is not stationary!

    Transformation of time series data to stationary by differencing techinque.

```python
# Apply first-order differencing on our data and store the result in widget_sales_diff.

widget_sales_diff = np.diff(df['widget_sales'], n=1)

# Add 'widget_sales_diff' to df_diff Data Frame

df_diff = pd.DataFrame({'widget_sales_diff': widget_sales_diff})
```

```python
sns.set(style="whitegrid")

# Create the figure and axis
fig, ax = plt.subplots(figsize=(10, 5))

# Plot the data using Seaborn's lineplot
sns.lineplot(data=df_diff, x=df_diff.index, y='widget_sales_diff', ax=ax)

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Widget sales ($1000)')
ax.set_title('Widget Sales Over Time')

# Customize the x-axis tick positions and labels
x_ticks = [
    0, 30, 57, 87, 116, 145, 175, 204, 234, 264, 293, 323, 352, 382, 409, 439, 468, 498
]
x_labels = [
    'Jan 2019', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec',
    'Jan 2020', 'Feb', 'Mar', 'Apr', 'May', 'Jun'
]
ax.set_xticks(x_ticks)
ax.set_xticklabels(x_labels, rotation=45, ha="right")

# Ensure a tight layout
plt.tight_layout()

plt.show()
```

```python
df_diff['widget_sales_diff']
```

```python
ADF_result = adfuller(df_diff['widget_sales_diff'])

print(f'ADF Statistic: {ADF_result[0]}')
print(f'p-value: {ADF_result[1]}')
```

    Inference:
        ADF statistic is large negative number, and the p-value is very small than 0.05.
    
        Hence : time series is stationary!

    Now we need to plot ACF plot which determine the order of a moving average process.

```python
from statsmodels.graphics.tsaplots import plot_acf

#Plot the ACF of the differenced series.
plot_acf(df_diff['widget_sales_diff'], lags=30);

plt.tight_layout()
```

    Inference:
        Notice how the coefficients are significant up until lag 2, and then they fall abruptly into the non-significance zone (shaded area) of the plot. 
        There are some significant coefficients around lag 20, but this is likely due to chance, since they are non-significant between lags 3 and 20 and after lag 20.

```python
# Spliting the data into train and test data

train = df_diff[:int(0.9*len(df_diff))]
test = df_diff[int(0.9*len(df_diff)):]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

def rolling_forecast(df: pd.DataFrame, train_len: int, horizon: int, window: int, method: str) -> list:
    
    total_len = train_len + horizon
    
    if method == 'mean':
        pred_mean = []
        
        for i in range(train_len, total_len, window):
            mean = np.mean(df[:i].values)
            pred_mean.extend(mean for _ in range(window))

        return pred_mean

    elif method == 'last':
        pred_last_value = []
        
        for i in range(train_len, total_len, window):
            last_value = df[:i].iloc[-1].values[0]
            pred_last_value.extend(last_value for _ in range(window))
            
        return pred_last_value
    
    elif method == 'MA':
        pred_MA = []
        
        for i in range(train_len, total_len, window):
            model = SARIMAX(df[:i], order=(0,0,2))
            res = model.fit(disp=False)
            predictions = res.get_prediction(0, i + window - 1)
            oos_pred = predictions.predicted_mean.iloc[-window:]
            pred_MA.extend(oos_pred)
            
        return pred_MA
```

```python
pred_df = test.copy()

TRAIN_LEN = len(train)
HORIZON = len(test)
WINDOW = 2

pred_mean = rolling_forecast(df_diff, TRAIN_LEN, HORIZON, WINDOW, 'mean')
pred_last_value = rolling_forecast(df_diff, TRAIN_LEN, HORIZON, WINDOW, 'last')
pred_MA = rolling_forecast(df_diff, TRAIN_LEN, HORIZON, WINDOW, 'MA')

pred_df['pred_mean'] = pred_mean
pred_df['pred_last_value'] = pred_last_value
pred_df['pred_MA'] = pred_MA

pred_df.head()
```

```python
# Set Seaborn style
sns.set(style="whitegrid")

# Create the figure and axis using Seaborn's lineplot
fig, ax = plt.subplots(figsize=(10, 6))

# Create line plots for different series using Seaborn
sns.lineplot(data=df_diff, x=df_diff.index, y='widget_sales_diff', ax=ax, label='Diff')
sns.lineplot(data=pred_df, x=pred_df.index, y='widget_sales_diff', ax=ax, label='Actual', linestyle='-', color='blue')
sns.lineplot(data=pred_df, x=pred_df.index, y='pred_mean', ax=ax, label='Mean', linestyle=':', color='green')
sns.lineplot(data=pred_df, x=pred_df.index, y='pred_last_value', ax=ax, label='Last', linestyle='-.', color='red')
sns.lineplot(data=pred_df, x=pred_df.index, y='pred_MA', ax=ax, label='MA(2)', linestyle='--', color='black')

# Add the axvspan using Matplotlib
ax.axvspan(449, 498, color='#808080', alpha=0.2)

# Set labels and title
ax.set_xlabel('Time')
ax.set_ylabel('Widget sales - diff (k$)')
ax.set_xlim(430, 500)

# Customize x-axis ticks
ax.set_xticks([439, 468, 498])
ax.set_xticklabels(['Apr', 'May', 'Jun'])

# Ensure a tight layout
plt.tight_layout()

# Display the legend
ax.legend(loc=2)

# Show the plot
plt.show()
```

```python
from sklearn.metrics import mean_squared_error

mse_mean = mean_squared_error(pred_df['widget_sales_diff'], pred_df['pred_mean'])

mse_last = mean_squared_error(pred_df['widget_sales_diff'], pred_df['pred_last_value'])

mse_MA = mean_squared_error(pred_df['widget_sales_diff'], pred_df['pred_MA'])

# Data for the bar plot
x = ['mean', 'last_value', 'MA(2)']
y = [mse_mean, mse_last, mse_MA]

# Set Seaborn style
sns.set(style="whitegrid")

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 6))

# Create the bar plot using Seaborn's sns.barplot
sns.barplot(x=x, y=y, ax=ax, palette='Blues')

# Set labels and limits
ax.set_xlabel('Methods')
ax.set_ylabel('MSE')
ax.set_ylim(0, 5)

# Add text labels for each bar
for index, value in enumerate(y):
    ax.text(x=index, y=value + 0.25, s=str(round(value, 2)), ha='center')

# Ensure a tight layout
plt.tight_layout()

# Show the plot (optional)
plt.show()
```
