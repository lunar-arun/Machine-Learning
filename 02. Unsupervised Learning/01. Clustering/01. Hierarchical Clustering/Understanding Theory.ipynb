# Hierarchical Clustering

    This algorithm groups similar data points into clusters.
    
    This algorithm builds an hierarchy of clusters, starting with all data points that assigned a cluster of its own.
    
    Here two nearest clusters are merged into same cluster.

    Two implementation of this algorithm :
        
        Bottom-Up -> Agglomerative Clustering
        
        Top-Down -> Divisive Clustering

    Both of this methods use distance metrics for identifying closeness of two clusters 
        
        Euclidean Distance
        
        Manhattan Distance
        
        Maximum Distance
        
        Mahalanobis Distance

    Euclidean Distance
    
Euclidean distance is the straight-line distance between two points in Euclidean space. 

For two points $ \text{p} = (p_1, p_2,..., p_n) \text{ and q} = (q_1, q_2,...., q_n)$, the Euclidean distance is calculated as :

Euclidean Distance = $ \sqrt{\sum_{i=1}^{n} (p_i - q_i)^2} $


    Manhattan Distance

Manhattan distance, also known as city block distance or taxicab distance, measures the distance between two points by the sum of the absolute differences of their coordinates. 

For two points $ \text{p} = (p_1, p_2,..., p_n) \text{ and q} = (q_1, q_2,...., q_n)$  the Manhattan distance is calculated as :

Manhattan Distance = $ \sum_{i=1}^{n} |p_i - q_i| $

    Maximum Distance (Chebyshev Distance)
    
Maximum distance, also known as Chebyshev distance, calculates the maximum absolute difference between the coordinates of two points along any dimension.

For two points $ \text{p} = (p_1, p_2,..., p_n) \text{ and q} = (q_1, q_2,...., q_n)$, the Maximum distance is calculated as :

Maximum Distance = $ \max_{i} |p_i - q_i| $

    Mahalanobis Distance
    
Mahalanobis distance measures the distance between a point and a distribution. It takes into account the variability of the data by considering the covariance matrix of the dataset.

For a point x with mean  and covariance matrix C, the Mahalanobis distance is calculated as:

Mahalanobis Distance = $ \sqrt{(\textbf{x} - \boldsymbol{\mu})^T \textbf{C}^{-1} (\textbf{x} - \boldsymbol{\mu})} $


## Agglomerative Clustering

    Bottom-Up Approach :
        Each data points are consider as clusters and nearest pints are clubbed together.
        
    Step-1 : Consider data points as separate clusters.(Own cluster)
    
    Step-2 : Compute pairwise distance for each data points using distance metrics.
    
    Step-3 : Merge the two closest clusters or data points based on the chosen distance measure.
    
    Step-4 : Recalculate the distances between the newly formed cluster and all other clusters or data points.
    
    Step-5 : Repeat steps 3 and 4 until only a single cluster remains.
    
    Generate a Dendrogram to identify the clusters.

    To find optimal number of cluster : using "index nos".
